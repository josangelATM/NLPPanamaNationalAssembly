```python
#Funcion para mostrar entidades. (NER)
def show_ents(doc):
    if doc.ents:
        for ent in doc.ents:
            print(ent.text+' - '+ent.label_+' - '+str(spacy.explain(ent.label_)))
    else:
        print('No named entities found.')
```

```python
#Como saber si una palabra está fuera del vocab (token.is_oov) y si tiene vector (has_vector)
tokens = nlp(u'Sala Ministro Ministerio Presidator Concejal')

for token in tokens:
    print(token.text, token.has_vector, token.vector_norm, token.is_oov)
```

```python
#Code para realizar  operaciones aritmeticas con Vectores tipo ()
from scipy import spatial

cosine_similarity = lambda x, y: 1 - spatial.distance.cosine(x, y)

ministro = nlp.vocab['ministro'].vector
hombre = nlp.vocab['hombre'].vector
mujer = nlp.vocab['mujer'].vector

# Now we find the closest vector in the vocabulary to the result of "hombre" - "woman" + "queen"
new_vector = ministro - hombre + mujer
computed_similarities = []

for word in nlp.vocab:
    # Ignore words without vectors and mixed-case words:
    if word.has_vector:
        if word.is_lower:
            if word.is_alpha:
                similarity = cosine_similarity(new_vector, word.vector)
                computed_similarities.append((word, similarity))

computed_similarities = sorted(computed_similarities, key=lambda item: -item[1])

print([w[0].text for w in computed_similarities[:10]])
```
